```python
from zipfile import ZipFile
import os
import pandas as pd
```

```python
data_folder_path = '/content/Iot_train/train'
```

```python
csv_files = [file for file in os.listdir(data_folder_path) if file.endswith('.csv')]
alldata = [pd.read_csv(os.path.join(data_folder_path, file)) for file in csv_files]
```

```python
df = []
```

```python
for file in csv_files:
  print(file)
  read = pd.read_csv(os.path.join(data_folder_path, file))
  df.append(read)
```

```python
alldf = pd.concat(df, axis=0, ignore_index=True)
```

```python
alldf
```

```python
for index, row in alldf.iterrows():
    if row.isna().sum() > 5:
      print(index)
      alldf.drop(index, inplace=True)
alldf.reset_index(drop=True, inplace=True)
```

```python
alldf
```

```python
alldf['time'] = alldf['time'].str[:8]
alldf[['time1', 'time2', 'time3']] = alldf['time'].str.split(':', expand=True)
alldf = alldf.drop('time', axis=1)
alldf = alldf.drop('index', axis=1)
alldf[['time1', 'time2', 'time3']] = alldf[['time1', 'time2', 'time3']].astype(float)
```

```python
alldf['time'] = alldf['time1']*3600 + alldf['time2']*60 + alldf['time3']
```

```python
alldf = alldf.drop('time1', axis=1)
alldf = alldf.drop('time2', axis=1)
alldf = alldf.drop('time3', axis=1)
alldf = alldf.drop('11Rain Accumulate (mm)', axis=1)
alldf = alldf.drop('11Rain count', axis=1)
```

```python
alldf['light'] = alldf['1LightHigh'] + alldf['1LightLow']
alldf = alldf.drop('1LightHigh', axis=1)
alldf = alldf.drop('1LightLow', axis=1)
```

```python
column_names = alldf.columns.tolist()
column_names
```

```python
for i in range(1104):
  for column in column_names:
    if pd.isna(alldf.loc[i, column]):
      if i == 0:
        alldf.loc[i, column] = alldf.loc[i+1, column]
      else:
        alldf.loc[i, column] = (alldf.loc[i-1, column] + alldf.loc[i+1, column])/2
```

```python
for w in column_names:
  for q in alldf.index:
    s = q
    e = q
    while pd.isna(alldf.loc[e, w]):
      if pd.isna(alldf.loc[s+1, w])== False:
        alldf.loc[e, w] = (alldf.loc[e-1, w]+ alldf.loc[s, w])/2
      s=s+1
```

```python
import numpy as np
alldf['time'] = np.sin((alldf['time']*628)/8640000)
```

```python
for column in column_names:
  if column != '23SoilWC ค.ชื้นดิน':
    nordf[column] = (nordf[column] - nordf[column].min()) / (nordf[column].max() - nordf[column].min())
```
